```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt
```

```python
office_df = pd.read_csv("datasets/office_episodes.csv")
office_df.head()
```

```python
for i in range(1, 10):
    season_data = office_df[office_df['season'] == i]
    max_rat = season_data['scaled_ratings'].max()
    best_episodes = season_data[season_data['scaled_ratings'] == max_rat]
    print(best_episodes)
"""seasons = list(best_ep.keys())
episodes = [df['episode_nu'].iloc[0] for df in best_ep.values()]
plt.scatter(seasons, episodes)
plt.show()"""

```

```python
colors = []

for lab, row in office_df.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append("red")
    elif 0.25 <= row['scaled_ratings'] < 0.50:
        colors.append("orange")
    elif 0.50 <= row['scaled_ratings'] < 0.75:
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")
        
colors[:5]
```

```python
sizes = []

for lab, row in office_df.iterrows():
    if row['has_guests'] == True:
        sizes.append(250)
    else:
        sizes.append(25)

sizes[:5]
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(24, 24))
plt.scatter(office_df["episode_number"], office_df["viewership_mil"], c = colors, s = sizes)

plt.title('Popularity, Quality, and Guest Appearances on the Office', size = 16)

plt.xlabel('Episode Number', size = 14)
plt.ylabel('Viewership (Millions)', size = 14)

plt.show()
```
